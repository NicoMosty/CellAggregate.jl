```julia
using DelimitedFiles
using Plots
```

```julia
data      = Vector()
data_path = Vector()
for i in filter(isdir, readdir())
    global rows      = Vector()
    global rows_path = Vector()
    for j in filter(isdir, readdir(i, join=true))
        val = filter(x->endswith(x, ".nw.dat"),readdir(j, join=true))[1]
        push!(rows,readdlm(val)[5:end,1:3])
        push!(rows_path,[split(j,"/")...])
    end
    push!(data,rows)
    push!(data_path,rows_path)
end
```

# Plots

## neck

```julia
for i in 1:size(data,1)-1
    display(plot(
        data[i][1][:,1],
        map(x -> x[:,2], data[i]),

        title = "$(data_path[i][1][1])",
        label=reshape(map(x -> x[2], data_path[i]),(1,size(data_path[i],1))),
        seriestype=:scatter,
        ylims=(0,0.5)
    ))
end
```

## width

```julia
for i in 1:size(data,1)-1
    display(plot(
        data[i][1][:,1],
        map(x -> x[:,3], data[i]),

        title = "$(data_path[i][1][1])",
        label=reshape(map(x -> x[2], data_path[i]),(1,size(data_path[i],1))),
        seriestype=:scatter,
        ylims=(0.7,1.1)
    ))
end
```
